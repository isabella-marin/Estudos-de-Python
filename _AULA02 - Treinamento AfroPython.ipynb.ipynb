```
from sklearn.datasets import load_iris #O sklearn já possui o iris no seu pacote, o que torna nosso trabalho bem mais simples, né?

iris = load_iris() 
```

```
type(iris) #ainda não está como dataframe
```

```
print(iris.feature_names) #nome das colunas dentro desse dataset, disponiveis
```

```
import pandas as pd 
import numpy as np  

df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],  #O  pd.DataFrame cria um pandas dataframe e o np.c_ concatena arrays
                  columns= iris['feature_names'] + ['Species'])
```

```
#escluindo linhas 
"""
O código abaixo exclui a linha de index 0 do nosso dataframe. 
O axis=0 significa que quero excluir uma linha de index '0'. 
O inplace=False significa 
que eu NÃO quero excluir essa linha do dataframe original.
ele só faz aparecer sem ela de forma temporario
"""
df.drop(0, axis=0, inplace=False) 
```

```
"""
O código abaixo exclui a coluna de nome do nosso dataframe. 
O axis=1 significa que quero excluir uma coluna chamada 'Species'. 
O inplace=False significa
que eu NÃO quero excluir essa coluna do dataframe original.
"""

df.drop(['Species'], axis=1, inplace=False) 
```

```
#amostra do dataframe, retorna N linhas aleatorios do df
#random_state mantem os mesmo numeros
#n= numero sorteado
#axis= 0 refere-se a linhas, se quisesse colunas seria 1
df.sample(n = 5, axis = 0, random_state = 10)
```
