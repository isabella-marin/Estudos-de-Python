```
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np 

# %matplotlib inline # se necessario configura a biblioteca para gerar saída dentro do frontend, alternativa ao plt.show()
```

Figure: corresponde a toda o gráfico, pode conter um ou mais eixos (plotagens). Você pode pensar em uma Figura como uma tela que contém plotagens com axes, gráficos, texto e rótulos.

instância da classe plt.Figure
Axes: É o que geralmente consideramos uma plotagem (plot). Uma caixa delimitadora com marcas e rótulos, que eventualmente conterão os elementos do gráfico que compõem a visualização. Ele contém dois ou três (no caso de 3D) objetos de eixo (axis).

tem um título, um rótulo x e um rótulo y.
instância da classe plt.Axes
Axis: Os eixos são as retas numeradas como objetos e cuidam de gerar os limites numéricos do gráfico.

Artist: Tudo o que se pode ver na figura é um artist, como objetos de texto, objetos Line2D, objetos de coleção. A maioria dos artists está ligada a eixos.

```
# GRÁFICO SIMPLES COM DOIS EIXOS
fig1 = plt.figure() # cria um objeto figura 1
ax0 = fig1.add_axes([0,0,1,1]) # add_axes - cria eixos horizontal e vertical
ax0.set_xlabel('Eixo X')
ax0.set_ylabel('Eixo Y')
ax0.set_title('Título do Gráfico') # seta o título

# apresenta a figura, necesário em notebooks executados localmente
plt.show()
```

```
# PLOTANDO DADOS

#subplot forma quantidade de linhas, colunas e indice do painel

# cria um espaço linear de 0 a 10 com 100 elementos
x = np.linspace(0, 10, 100)
print(x)

# cria uma figura para plotagem
fig2 = plt.figure()

# crie o primeiro de dois painéis e defina o eixo atual
plt.subplot(2, 1, 1) # (linhas, colunas, numero do painel)
plt.plot(x, np.sin(x)) # x, seno de x

# cria o segundo painel e seta o eixo corrente
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x)); # x, cosseno de x
```

```
x = np.linspace(0, 10, 100)

# Primeiro cria uma grade de plotagens
fig, ax = plt.subplots(2)

# chama o método plot( x, f(y) ) dos objetos do array

# plota função seno no primeiro quadro
ax[0].plot(x, np.sin(x)) 

# plota a função cosseno no segundo quadro
ax[1].plot(x, np.cos(x), '--'); 
```

```
#multiplas linhas
fig, ax = plt.subplots(1)
ax.plot(x, np.sin(x)) 
ax.plot(x, np.cos(x), '--'); 
```

```
plt.plot(x, np.sin(x - 0), color='blue')        # cor pelo nome
plt.plot(x, np.sin(x - 1)/2, color='y')           # código curto de cor (rgbcmyk)
plt.plot(x, np.sin(x - 2)/3, color='0.75')        # escala de cinza entre 0 e 1
plt.plot(x, np.sin(x - 3)/4, color='#FFDD44')     # cores em hexadecinal (RRGGBB de 00 a FF)
plt.plot(x, np.sin(x - 4)/5, color=(1.0,0.2,0.3)) # tupla RGB tuple, valores de 0 a 1
plt.plot(x, np.sin(x - 5)/6, color='coral'); # todos os nomes de cores HTML são suportados
```

```
plt.plot(x, x + 0, linestyle='solid')  # solida
plt.plot(x, x + 1, linestyle='dashed') # tracejada
plt.plot(x, x + 2, linestyle='dashdot')# traço ponto
plt.plot(x, x + 3, linestyle='dotted') # pontos

# De forma mais curta, pode ser usado os seguintes códigos:
plt.plot(x, x + 4, linestyle='-')  # solida
plt.plot(x, x + 5, linestyle='--') # tracejada
plt.plot(x, x + 6, linestyle='-.') # traço ponto
plt.plot(x, x + 7, linestyle=':')  # pontos
```

```
plt.plot(x, x + 0, '-g')  # solida verde
plt.plot(x, x + 1, '--c') # tracejada ciano
plt.plot(x, x + 2, '-.k') # traço ponto preto
plt.plot(x, x + 3, ':r');  # pontilhada vermelha
```

```
#inverte a origem
plt.plot(x, np.sin(x))

plt.xlim(15, 0)
plt.ylim(-2, 5);
```

```
plt.plot(x, np.sin(x))
plt.axis([-5, 15, -2, 5]);
```

```
#'apertar o grafico, compacto'
plt.plot(x, np.sin(x))
plt.axis('tight');
```

```
#auto ajusta o tamanho
plt.plot(x, np.sin(x))
plt.axis('equal');
```

```
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.plot(x, np.tan(x), '--r', label='tan(x)')

#plt.axis('equal');

# ajustas os limites do gráfico
plt.axis([-1, 11, -45, 20]) 

plt.legend();
```

```
x = [0, 1, 2, 3, 4]
y1 = [4, 2, 9, 7, 1]
y2 = [5, 1, 7, 3, 8]

# aqui usamos a OO mas ax não é um array (figura de apenas 1 painel)
fig, ax = plt.subplots()

ax.plot(x, y1, '-b', label='linha A')
ax.plot(x, y2, '--r', label='linha B')

ax.axis('tight') # aperta
ax.xaxis.tick_top() # apresenta os marcadores do eixo X no topo

leg = ax.legend(); # mostra a legenda
# objeto para manipular a leganda
leg.set_title('Título da Legenda');

type(leg)
```
